```python
#dictionary json
top100 = {}
#scanning top 100 companies
for i in range(1,101):
    top100[i]={
        "name":table.find_all('tr')[i].find_all('td')[0].find_all('a')[0].get_text(),
        "website":table.find_all('tr')[i].find_all('td')[0].find_all('a')[1].get_text(),
        "rank":table.find_all('tr')[i].find_all('td')[1].get_text()[1:len(table.find_all('tr')[i].find_all('td')[1].get_text())],
        "description":table.find_all('tr')[i].find_all('td')[2].get_text(),
        "status":table.find_all('tr')[i].find_all('td')[3].get_text(),
        "batch":table.find_all('tr')[i].find_all('td')[4].get_text(),
        "hq":table.find_all('tr')[i].find_all('td')[6].get_text()
    }
    try:
        if (table.find_all('tr')[i].find_all('td')[0].img['src']!=None):
            top100[i]["logo"]=table.find_all('tr')[i].find_all('td')[0].img['src']
    except:
        print("exception occured")
    try:
        if (table.find_all('tr')[1].find_all('td')[7].a['href']!=None):
            top100[i]['apply']=table.find_all('tr')[1].find_all('td')[7].a['href']
    except:
        print("exception occured")
    
    if(table.find_all('tr')[i].find_all('td')[5].get_text()==''):
        top100[i]["breakthrough"]=1
    else:
        top100[i]["breakthrough"]=0

    if(';' in table.find_all('tr')[i].find_all('td')[6].get_text()):
        top100[i]["hq"]=table.find_all('tr')[i].find_all('td')[6].get_text().split(";")
    else:
        top100[i]["hq"]=table.find_all('tr')[i].find_all('td')[6].get_text()

    
```
